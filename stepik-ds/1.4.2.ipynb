```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test/iris.csv', index_col=0)
```

```python
df.head()
```

```python
for column in df:
    ax = sns.kdeplot(df[column], label=column)
ax.legend()
```

```python
for column in df:
    bx = sns.distplot(df[column], label=column)
bx.legend()
```

```python
sns.violinplot(y=df['petal length'])
```

```python
sns.pairplot(df, hue='species')
```
